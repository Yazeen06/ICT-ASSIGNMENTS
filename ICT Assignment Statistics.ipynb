```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbs
                
```

```python
db=pd.read_csv('C:/Users/Dell/Downloads/StudentsPerformance.csv')
```

```python
db
```

# Males and Females participated in the test.

```python
c=db.value_counts('gender')
print('Number of Females participated in the test: {}'.format(c))
```

# Students parental level of education

```python
d=db.value_counts('parental level of education')
y=d.plot(kind='bar',figsize=(16,8),color=['Yellow','red','green','blue','violet'])
plt.xticks(rotation=360);
plt.title('Parental Level Of Education')
plt.ylabel('Count of Parents')
for p in y.patches: 
        y.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
```

# Who scores the most on average for math, reading and writing based on

# Gender

```python
g=db[['gender','math score','reading score','writing score']]
gg=g.groupby('gender').mean().transpose()
gg.plot(kind='bar',figsize=(12,8))
plt.xticks(rotation=360);
plt.title('Average for Math.Reading and Writing',fontsize=20);
```

# Test preparation course

```python
tp=db[['test preparation course','math score','reading score','writing score']]
tp1=tp.groupby('test preparation course').mean().transpose()
tp1.plot(kind='bar',figsize=(12,8),color=['green','yellow'])
plt.xticks(rotation=360)
```

# scoring variation for math, reading and writing based on

# Gender

```python
db
```

```python
s=db[['gender','math score','reading score','writing score']]
s1=s.groupby('gender').std().transpose().plot(kind='bar',figsize=(12,8),color=['grey','green'])
plt.xticks(rotation=360);
for p in s1.patches: 
        s1.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
```

# Test Preparation Course

```python
s=db[['test preparation course','math score','reading score','writing score']]
s1=s.groupby('test preparation course').std().transpose().plot(kind='line',figsize=(12,8),color=['red','blue'])
plt.xticks(rotation=360);
for p in s1.patches: 
        s1.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
```
